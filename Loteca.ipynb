```python
import pandas as pd
import numpy as np
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("BRA.csv")
```

```python
vocab = df['Home'].unique()
```

```python
word2idx = {word: i for i, word in enumerate(vocab)}
```

```python
def text_to_vector2(text): 
    text_vector = text.split(',')
    word_vector = np.zeros(len(text_vector), dtype=np.int_)
    for j in range(len(text_vector)):
        idx = word2idx.get(text_vector[j], None)
        if idx is None:
            continue
        else:
            word_vector[j] = idx
    return np.array(word_vector)
```

```python
word_vectors2 = []
for _, row in df.iterrows():
    word_vectors2.append(text_to_vector2(row['Home'] + ',' + row['Away']))
    row['iHome'] = text_to_vector2(row['Home'] + ',' + row['Away'])
```
