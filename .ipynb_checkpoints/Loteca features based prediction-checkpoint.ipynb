# Loteca

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot
```

### Ler arquivo

```python
df = pd.read_csv("BRA.csv")
```

```python
df.dropna()
```

```python
df.head()
```

```python
X_all = df.copy()
```

```python
# Standardising the data.
from sklearn.preprocessing import scale

#Center to the mean and component wise scale to unit variance.
cols = [['PH','PD','PA','MaxH','MaxD','MaxA','AvgH','AvgD','AvgA']]
for col in cols:
    X_all[col] = scale(X_all[col])
```

```python
X_all.dropna()
```

```python
y_all = X_all['Res']
```

```python
y = []
for l in y_all:
    if l == 'H':
        y.append(0)
    else:
        if l == 'D':
            y.append(1)
        else:                
            if l == 'A':
                y.append(2)
            else:
                y.append(3)
```

### Separar os datasets de treinamento e teste

```python
X_train, X_test, y_train, y_test = train_test_split(X_all[['PH','PD','PA','MaxH','MaxD','MaxA','AvgH','AvgD','AvgA']], y, test_size=0.33, random_state=42, shuffle=True)
```

### Suport vector machine

```python
from sklearn.svm import SVC
model = SVC(gamma='scale', C=1.0)
```

### Random forest

```python
# from sklearn.ensemble import RandomForestRegressor
# from sklearn.datasets import make_regression
# model = RandomForestRegressor(max_depth=2, random_state=0, n_estimators=100)
```

### Treinar o modelo

```python
model.fit(X_train, y_train)
```

```python
model.predict([X_test.values[11]])
```

### Avaliar precisão

```python
y_pred = []
for X_values in X_test.values[:-5]:
    y_pred.append(model.predict([X_values]))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test[:-5], y_pred)
```
